```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
predicted_data = pd.read_csv("./datasets/predicted.csv")
predicted_data.head()
important_features = ["plan_effective_months","contract_type","income","medium_risk","debit_code","high_risk","use_client_address","growth","speculation","inception_months","type_code", "label"]
predicted_data = predicted_data[important_features]
predicted_data.head()
```

```python

#     if column != 'label':
#         sns.countplot(x = column, hue = "label", data=predicted_data)
for column in predicted_data.columns:
    if column != 'label':
        sns.boxplot(x='label', y=column, data=predicted_data)
        plt.title(f'Distribution of {column} for Churn and Non-Churn Instances')
        plt.show()
```
